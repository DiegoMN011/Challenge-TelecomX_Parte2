Importar biblioteca.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

Carregar dados tratados.

```python
dados = pd.read_csv("dados_tratados.csv")

dados.head()
```

Verificar Estrutura de dados.

```python
dados.info()
```

Remover coluna irrelevante.

```python
dados = dados.drop(columns=["customerID"])
```

Encoding

```python
dados = pd.get_dummies(dados, drop_first=True)

dados.head()
```

Verificar proporção de churn

```python
print(dados['Churn_Yes'].value_counts())

sns.countplot(x='Churn_Yes', data=dados)

plt.title("Distribuição de Evasão de Clientes")
plt.show()
```

Matriz de correlação

```python
plt.figure(figsize=(15,10))

sns.heatmap(dados.corr(), cmap="coolwarm")

plt.title("Matriz de Correlação")
plt.show()
```

Análises direcionadas

```python
Tempo de cliente vs churn
sns.boxplot(x="Churn_Yes", y="customer_tenure", data=dados)

plt.title("Tempo de Permanência vs Evasão")
plt.show()
Gasto mensal vs churn
sns.boxplot(x="Churn_Yes", y="account_Charges_Monthly", data=dados)

plt.title("Gasto Mensal vs Evasão")
plt.show()
```

Separar variáveis

```python
X = dados.drop("Churn_Yes", axis=1)

y = dados["Churn_Yes"]
```

Separar treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

Normalização (para regressão logística)

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
🧩 CÉLULA 12 — Modelo 1 (Regressão Logística)
modelo_lr = LogisticRegression(max_iter=1000)

modelo_lr.fit(X_train_scaled, y_train)

pred_lr = modelo_lr.predict(X_test_scaled)
```

Modelo 2 (Random Forest)

```python
modelo_rf = RandomForestClassifier(random_state=42)

modelo_rf.fit(X_train, y_train)

pred_rf = modelo_rf.predict(X_test)
```

Função de avaliação

```python
def avaliar_modelo(y_test, pred):

    print("Accuracy:", accuracy_score(y_test, pred))
    print("Precision:", precision_score(y_test, pred))
    print("Recall:", recall_score(y_test, pred))
    print("F1-score:", f1_score(y_test, pred))

    print("\nRelatório de Classificação")
    print(classification_report(y_test, pred))

    matriz = confusion_matrix(y_test, pred)

    sns.heatmap(matriz, annot=True, fmt="d", cmap="Blues")

    plt.title("Matriz de Confusão")
    plt.show()
```

Avaliar regressão logística

```python
avaliar_modelo(y_test, pred_lr)
```

Avaliar random forest

```python
avaliar_modelo(y_test, pred_rf)
```

Importância das variáveis

```python
importancias = modelo_rf.feature_importances_

variaveis = pd.Series(importancias, index=X.columns)

variaveis.sort_values(ascending=False).head(10)
```

Gráfico das variáveis mais importantes

```python
variaveis.sort_values(ascending=False).head(10).plot(kind="bar")

plt.title("Variáveis mais importantes para prever Churn")

plt.show()
```

Conclusão

Conclusão

O desenvolvimento dos modelos preditivos permitiu identificar fatores importantes relacionados à evasão de clientes da Telecom X.

Entre as variáveis mais relevantes estão o tipo de contrato, o tempo de permanência do cliente e os valores cobrados mensalmente.

Clientes com contratos mensais e menor tempo de permanência apresentam maior probabilidade de cancelar o serviço.

Entre os modelos avaliados, o Random Forest apresentou melhor desempenho geral, demonstrando maior capacidade de capturar padrões complexos nos dados.

Essas informações podem ajudar a empresa a identificar clientes com maior risco de evasão e desenvolver estratégias de retenção mais eficazes.
